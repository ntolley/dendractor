```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_samples = 1000
len_stim = 100
context_start = 200
cue_start = 700

context_end = context_start + len_stim
cue_end = cue_start + len_stim

# Declare task vars
t = np.linspace(0, 1, 1000)
u0 = np.zeros_like(t)
u1 = np.zeros_like(t)
u2 = np.zeros_like(t)
z1 = np.zeros_like(t)
z2 = np.zeros_like(t)

# Assign task values
u0[context_start:context_end] = 1
u1[cue_start:cue_end] = 1
u2[cue_start:cue_end] = -1
z1[cue_start:] = 1
z2[cue_start:] = -1

# MemoryAnti condition
u0_anti = u0 * -1
u1_anti = u1
u2_anti = u2
z1_anti = z1 * -1
z2_anti = z2 * -1
```

```python
def plot_fill(x, y, offset, color='C0', label='$u_0$', alpha=0.3):
    plt.plot(x, y + offset, color=color)
    plt.fill_between(x, np.zeros_like(y) + offset, y + offset, color=color, alpha=alpha)
    plt.text(-0.2, offset, label, fontsize=16)

def plot_time_axis(offset):
    offset = offset + 0.5
    plt.plot(t, np.zeros_like(t) + offset, color='k', alpha=1.0)
    plt.scatter(t[-1], offset + 0.01, marker=5, s=60, color='k', alpha=1.0)
    plt.text(0, -1, 'time', fontsize=12)
```

```python
color_dict = {'u0': 'C0', 'u_cue': 'C2', 'z': 'C3'}

y_pos = np.linspace(0, 13, 6)[::-1]
y_min, y_max= -2, 15

plt.figure(figsize=(7,4.5))
plt.subplot(1,2,1)
plot_fill(t, u0, offset=y_pos[0], color=color_dict['u0'], label='$u_0$')
plot_fill(t, u1, offset=y_pos[1], color=color_dict['u_cue'], label='$u_1$')
plot_fill(t, u2, offset=y_pos[2], color=color_dict['u_cue'], label='$u_2$')
plot_fill(t, z1, offset=y_pos[3], color=color_dict['z'], alpha=0.0, label='$\\hat{z}_1$')
plot_fill(t, z2, offset=y_pos[4], color=color_dict['z'], alpha=0.0, label='$\\hat{z}_2$')
plot_time_axis(y_pos[5])
plt.text(0.2, y_max, 'MemoryPro', fontsize=16)
plt.ylim(y_min, y_max)


plt.axis('off')

plt.subplot(1,2,2)
plot_fill(t, u0_anti, offset=y_pos[0], color=color_dict['u0'], label='$u_0$')
plot_fill(t, u1_anti, offset=y_pos[1], color=color_dict['u_cue'], label='$u_1$')
plot_fill(t, u2_anti, offset=y_pos[2], color=color_dict['u_cue'], label='$u_2$')
plot_fill(t, z1_anti, offset=y_pos[3], color=color_dict['z'], alpha=0.0, label='$\\hat{z}_1$')
plot_fill(t, z2_anti, offset=y_pos[4], color=color_dict['z'], alpha=0.0, label='$\\hat{z}_2$')
plot_time_axis(y_pos[5])
plt.text(0.2, y_max, 'MemoryAnti', fontsize=16)
plt.ylim(y_min, y_max)

plt.axis('off')
# plt.savefig('task_condition_schematic.svg')
```
